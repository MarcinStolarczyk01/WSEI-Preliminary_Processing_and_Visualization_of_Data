# Dataset: "Monthly average earth surface temperature"
[Dtaset Origin](https://ourworldindata.org/grapher/monthly-average-surface-temperatures-by-year)

```python
import pandas as pd

df = pd.read_csv("monthly-average-surface-temperatures-by-year.csv")
df = df.rename(columns={"Year": "Month"})
df.head(5)
```

```python
# pick average Word temperature
df = df[df.Entity == "World"]
# drop code and year 2025 (not complete data)
df = df.drop(columns=["Code", "2025"])
```

```python
df
```

```python
# change format from wide to long, apply datatime
import numpy as np
import datetime

# [Date, Temp]
rows = []

for i, r in df.iterrows():
    years_temp = r.iloc[2:]
    month = r[1]
    for year, temp in years_temp.items():
        date = datetime.datetime(year=int(year), month=month, day=1)
        row = [date, temp]
        rows.append(row)
```

```python
world_df = pd.DataFrame(rows, columns=["date", "temp"])
```

```python
world_df.to_csv("world_avg_temp.csv", index=False)
```
